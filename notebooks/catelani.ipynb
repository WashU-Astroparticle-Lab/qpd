We consider a low-energy effective Hamiltonian of the Josephson junction (excluding the resonator)

$$
\hat{H}=\hat{H}_{\varphi}+\hat{H}_{\mathrm{qp}}+\hat{H}_T
$$

The first term determines the dynamics of the phase degree of freedom in the absence of quasiparticles:

$$
\hat{H}_{\varphi}=4 E_C\left(\hat{N}-n_g\right)^2-E_J \cos \hat{\varphi}
$$

where $\hat{N}=-i d / d \varphi$ is the number operator of Cooper pairs passed across the junction, $n_g$ is the dimensionless gate voltage, and the parameters characterizing the qubit are charging energy $E_C$ and Josephson energy $E_J$.

The second term in the JJ Hamiltonian is the sum of the BCS Hamiltonians for quasiparticles in the leads

$$
\hat{H}_{\mathrm{qp}}=\sum_{j=L, R} \hat{H}_{\mathrm{qp}}^j, \quad \hat{H}_{\mathrm{qp}}^j=\sum_{n, \sigma} \epsilon_n^j \hat{\alpha}_{n \sigma}^{j \dagger} \hat{\alpha}_{n \sigma}^j,
$$

where $\hat{\alpha}_{n \sigma}^j\left(\hat{\alpha}_{n \sigma}^{j \dagger}\right)$ are quasiparticle annihilation (creation) operators and $\sigma=\uparrow, \downarrow$ accounts for spin. The quasiparticle energies are 

$$
\epsilon_n^j=\sqrt{\left(\xi_n^j\right)^2+\left(\Delta^j\right)^2}
$$

with $\xi_n^j$ and $\Delta_j$ being the single-particle energy level $n$ in the normal state of lead $j$ and the gap parametere in that lead, respectively. The occupations of the quasiparticle states are described by the distribution functions

$$
f^j\left(\xi_n^j\right)=\left\langle\hat{\alpha}_{n \sigma}^{j \dagger} \hat{\alpha}_{n \sigma}^j\right\rangle_{\mathrm{qr}}, \quad j=L, R
$$

assumed to be independent of spin; angular brackets denote averaging over the quasiparticle states. Hereinafter, we assume for simplicity equal gaps in the leads 

$$
\Delta^L=\Delta^R \equiv \Delta
$$

The last term in the qubit Hamiltonian describes quasiparticle tunneling across the junction and couples the phase and quasiparticle degrees of freedom:
$$
\hat{H}_T=\tilde{t} \sum_{n, m, \sigma}\left(e^{i \frac{\hat{\varphi}}{2}} u_n^L u_m^R-e^{-i \frac{\hat{\varphi}}{2}} v_m^R v_n^L\right) \hat{\alpha}_{n \sigma}^{L \dagger} \hat{\alpha}_{m \sigma}^R+\text { H.c. }
$$

where $u$ and $v$ are Bogoliubov amplitudes, representing the electron-like amplitude and the hole-like probability amplitude for a quasiparticle (which is a superposition of both). $\tilde{t}$ is the bare electron tunneling amplitude across the barrier. Note that the tunneling conductance for a standard metal-insulator-metal:

$$
g_T=\frac{4 \pi e^2}{\hbar} \nu^L \nu^R|\tilde{t}|^2
$$

In low-energy approximation for the qubit energy $\omega$ and characteristic energy $\delta E$ of quasiparticles are small compared to $\Delta$, we can further approximate by $u_m^j \simeq v_n^j \simeq 1 / \sqrt{2}$. This simplifies the equation to:

$$
\hat{H}_T=\tilde{t} \sum_{n, m, \sigma} i \sin \frac{\hat{\varphi}}{2} \hat{\alpha}_{n \alpha}^{L \dagger} \hat{\alpha}_{m \sigma}^R+\text { H.c. }
$$
